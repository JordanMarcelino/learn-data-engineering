## Simple ingestion with HTML scrapping with Bs4


```python
url = "https://openlibrary.org/trending/daily"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
books = []

for book in soup.find_all("li", class_="searchResultItem sri--w-main"):
    image = book.find("img", attrs={"itemprop": "image"})
    title = book.find("div", class_="details").find("a", class_="results")
    author = book.find("span", class_="bookauthor").find("a", class_="results")
    published = book.find("span", class_="publishedYear")

    books.append(
        {
            "image": image.get("src"),
            "title": title.text.strip(),
            "author": author.text.strip(),
            "published": published.text.strip(),
        }
    )
```

```python
df = pd.DataFrame(books)
```
